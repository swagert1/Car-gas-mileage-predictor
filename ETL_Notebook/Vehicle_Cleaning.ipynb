```python
import pandas as pd
import numpy as np
import requests as rq

import datetime as dt
import pandas as pd
```

```python
# load csv with vehicle EPA data

vehicles_df = pd.read_csv('../Resources/vehicles.csv', low_memory=False)
vehicles_df.head(3)
```

```python
# list columns

vehicles_df.columns.values
```

```python
# get number of alternative fuel vehicles in dataset

vc = vehicles_df.atvType.value_counts()
vc.sum()
```

```python
# removes all alternative fuel vehicles in data set including hybrids

vehicles_df = vehicles_df[vehicles_df.atvType.isnull()]
vehicles_df.head(3)
```

```python
# reduce columns to 8 for easier manipulation and viewing

veh_reduced = vehicles_df[['year','VClass','comb08','cylinders','displ','drive','fuelType','trany']]
veh_reduced
```

```python
#remove all rows with missing any data

veh_reduced.dropna(how = 'any', inplace = True)
veh_reduced
```

```python
veh_reduced.dtypes
```

```python
# sort data by year ascending
sorted = veh_reduced.sort_values(by='year', axis=0, ascending=True)
sorted
```

```python
#sorted.to_csv("../Resources/vehicles_basic_clean.csv",index=False)
```

```python
# getting counts of vehicle numbers

VClass_counts = sorted.VClass.value_counts()
VClass_counts.sort_index()
```

```python
# function to automate reductions

def clean_up(df, column, replace):
    
    for c in replace:
        df.loc[vr[column].str.contains(c), column] = c
        
```

```python
# reduce VClasses

vr = sorted.copy()
class_search_values = ["Minivan", "Small Pickup Trucks", "Special Purpose Vehicle", "Sport Utility Vehicle",\
                      "Standard Pickup Trucks","Vans", "Small Sport Utility Vehicle", "Standard Sport Utility Vehicle"]

# for c in class_search_values:
#     vr.loc[vr["VClass"].str.contains(c), "VClass"] = c
clean_up(vr, "VClass", class_search_values)    
    
vr.loc[vr["VClass"].str.contains("Midsize Station"), "VClass"] = "Midsize-Large Station Wagons"
    
vr["VClass"].value_counts()
```

```python
# reduce to 4, front or rear drive

vr.loc[vr["drive"].str.contains("2-Wheel Drive"), "drive"] = "Front-Wheel Drive"
# vr.loc[vr["drive"].str.contains("Rear-Wheel"), "drive"] = "2-Wheel Drive"
vr.loc[vr["drive"].str.contains("All-Wheel"), "drive"] = "4-Wheel Drive"
vr.loc[vr["drive"].str.contains("Part-time"), "drive"] = "4-Wheel Drive"
drive_counts = vr.drive.value_counts()
drive_counts.sort_index() 
```

```python
# reduce to automatic or manual

AorM = ["Automatic", "Manual"]

clean_up(vr, "trany", AorM )

trany_counts = vr.trany.value_counts()
trany_counts.sort_index()
```

```python
fuel_count = vr.fuelType.value_counts()
fuel_count
```

```python
# remove any remaining non-gasoline fuels

vr = vr[(vr.fuelType == "Regular") | (vr.fuelType == "Premium") | (vr.fuelType == "Midgrade")]
vr
```

```python
# remove fuelType from columns as no longer needed
vr = vr.drop("fuelType", axis=1)
```

```python
vr = vr.rename(columns={"comb08": "MPG(comb)", "VClass": "vehicle_class","displ":"displ (L)"})
vr
```

```python
# vr.to_csv("../Resources/vehicles_fully_reduced_2.csv",index=False)
```
